```python

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['text.usetex'] = False

#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
#mpl.rcParams['savefig.dpi'] = 1200
#mpl.rcParams['text.usetex'] = True  # not really needed
#plt.rc('text', usetex=True)
#plt.rc('font', family='serif')

fig, ax = plt.subplots(figsize=(8,6))

noise = np.array([0.1,  1, 5, 10])
rmse_5000 = np.array([0.1732, 0.1736, 0.1797, 0.2031])
rmse_10000 = np.array([0.1549, 0.1749, 0.1783, 0.1902])
rmse_15000 = np.array([0.1014, 0.1622, 0.1734, 0.1794])
c1 = "royalblue"
c2 = "green"
c3 = "red"
ax.scatter(noise, rmse_5000, s=s, color="royalblue" )
ax.plot(noise, rmse_5000, lw=lw, color="royalblue", label='m=5000')
ax.scatter(noise, rmse_10000, s=s, color=c2)
ax.plot(noise, rmse_10000, lw=lw, color=c2, label='m=10000')
ax.scatter(noise, rmse_15000, s=s, color=c3)
ax.plot(noise, rmse_15000, lw=lw, color=c3, label='m=15000')
plt.yticks(np.arange(0, 0.3, 0.02), fontsize=5)
plt.xticks([0.1,1, 5, 10], [0.1,1, 5, 10], fontsize=5)
plt.ylim(0.09, 0.21)
plt.xlim(0, 10.1)

plt.xlabel("Noise", labelpad=20, fontsize=label_size)
plt.ylabel("RMSE", fontsize=label_size, labelpad=20)
#ax.grid(True, lw=0.9)
ax2 = ax.twinx()
ax2.set_yticks(ax.get_yticks())
ax2.set_ylim(ax.get_ylim())
ax2.yaxis.set_tick_params(labelleft=False, labelright=False, direction='in' )
ax.tick_params(labelsize=label_size)
ax.legend(prop=prop)
plt.tight_layout()

plt.show()
#plt.savefig(f'./figures/correlation_w_densities.pdf', format='pdf', dpi=100)
```
