```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import rc
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
from matplotlib import ticker

mpl.rcParams.update(mpl.rcParamsDefault)

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
# title = r'$\mathrm{Synthetic~(2~obervations)}$'
d2 = [2000, 4000, 6000, 8000, 10000]

runtime_2 = [0.74291048, 0.821937, 0.87259, 0.93277, 1.0176]
runtime_6 = [0.80854, 0.86233, 0.91836, 0.98219, 1.09706]
runtime_10 = [1.301944, 1.3701, 1.437, 1.53194, 1.6501]
runtime_std = [0.00783107, 0.00995871, 0.02235781, 0.39976870, 0.0052526]

f, ax = plt.subplots(figsize=(8,7))

plt.plot(d2, runtime_10, lw=5, color='royalblue', ls="dashdot")
plt.scatter(d2, runtime_10, s=200, color='royalblue', label=r"$C=10$")

plt.plot(d2, runtime_6, lw=5, color='gold', ls="dotted")
plt.scatter(d2, runtime_6, s=200, color='gold', label=r"$C=6$")

plt.plot(d2, runtime_2, lw=5, color='red', ls="dashed")
plt.scatter(d2, runtime_2, s=200, color='red', label=r"$C=2$")


ax.legend(fontsize=32, loc='upper right', bbox_to_anchor=(0.55, 0.65))


plt.xlabel(r"$d$", fontsize=40)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.xaxis.get_offset_text().set_fontsize(40)

xticks = d2
plt.xticks(xticks, fontsize=40)
plt.yticks([0.8, 1.2, 1.6], fontsize=40)
plt.ylabel(r"$\mathrm{Time~(s)}$", fontsize=40)


plt.grid(lw=0.5, ls=":")
plt.tight_layout()
plot_name = f"runtime_syn.pdf"
plt.savefig(plot_name, bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```

```python
# title = r'$\mathrm{Synthetic~(2~obervations)}$'
m = [2e7, 2.5e7, 3.2e7]

runtime_8 = [48, 110, 150]
runtime_5 = [37, 77.3, 109]
runtime_2 = [12, 21, 32]

f, ax = plt.subplots(figsize=(8,7))

plt.plot(m, runtime_8, lw=5, color='royalblue', ls="dashdot")
plt.scatter(m, runtime_8, s=200, color='royalblue', label=r"$p=0.8$")

plt.plot(m, runtime_5, lw=5, color='gold', ls="dotted")
plt.scatter(m, runtime_5, s=200, color='gold', label=r"$p=0.5$")

plt.plot(m, runtime_2, lw=5, color='red', ls="dashed")
plt.scatter(m, runtime_2, s=200, color='red', label=r"$p=0.2$")

ax.legend(fontsize=32, loc='upper right', bbox_to_anchor=(0.51, 1.03))

plt.xlabel(r"$m$", fontsize=40)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.xaxis.get_offset_text().set_fontsize(40)

xticks = m
plt.xticks(xticks, fontsize=40)
plt.yticks([50, 100, 150], fontsize=40)
plt.ylim(0, 180)
plt.ylabel(r"$\mathrm{Time~(s)}$", fontsize=40)



plt.grid(lw=0.5, ls=":")
plt.tight_layout()
plot_name = f"runtime_real.pdf"
plt.savefig(plot_name, bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```
