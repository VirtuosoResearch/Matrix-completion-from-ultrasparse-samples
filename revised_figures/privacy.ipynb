```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import rc
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import ScalarFormatter
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
# title = r'$\mathrm{Synthetic~(2~obervations)}$'
noise_ratio = [0.01, 0.1, 0.2, 0.4, 0.6, 0.8, 1]

n5000 = [0.0070049, 0.0722105, 0.1427792, 0.2767241, 0.4116955, 0.5301328, 0.6483357]
n5000_std = [0.0000746, 0.0002172, 0.0002942, 0.0019634, 0.0025822, 0.0041475, 0.0035879]

n10000 = [0.0046358, 0.0488830, 0.0994144, 0.1956577, 0.2914413, 0.3791932, 0.4704019]
n10000_std = [0.0000592, 0.0003127, 0.0007505, 0.0005185, 0.0031290, 0.0029973, 0.0028909]

n20000 = [0.0033604, 0.0324980, 0.0659016, 0.1325967, 0.1990896, 0.2616272, 0.3249923]
n20000_std = [0.0000283, 0.0001585, 0.0009766, 0.0013878, 0.0023296, 0.0016603, 0.0036520]

f, ax = plt.subplots(figsize=(8,4))

#enhance = ob2_rmse_mean / rmse_mean
#enhance = rmse_mean / ob2_rmse_mean
#print(enhance.mean())
#print(ob2_err_mean)
#print(err_mean)

plt.plot(noise_ratio, n5000, lw=3, color='royalblue',ls="dashdot" )
plt.plot(noise_ratio, n10000, lw=3, color='gold', ls="dotted")
plt.plot(noise_ratio, n20000, lw=3, color='red', ls="dashed")

plt.scatter(noise_ratio, n5000, s=100, color='royalblue', label=r"$n=5000$", marker="X")
plt.scatter(noise_ratio, n10000, s=100, color='gold', label=r"$n=10000$")
plt.scatter(noise_ratio, n20000, s=100, color='red', label=r"$n=20000$", marker="D")

plt.errorbar(noise_ratio, n5000, yerr=n5000_std,color='royalblue', fmt='X', capsize=10)
plt.errorbar(noise_ratio, n10000, yerr=n10000_std,color='gold', capsize=10)
plt.errorbar(noise_ratio, n20000, yerr=n20000_std,color='red', fmt='D', capsize=10)

plt.legend(fontsize=26, loc='center left',bbox_to_anchor=(1, 0.5), ncol=1)

plt.xlabel(r"$\sigma$", fontsize=26)
plt.ylabel(r"$\mathrm{Estimation~Error}$", fontsize=26)


xticks = [0, 0.2, 0.4, 0.6, 0.8, 1]
plt.xticks(xticks, fontsize=26)
plt.ticklabel_format(style='sci', scilimits=(0,0), axis='x')
plt.yticks([0.1, 0.3, 0.5, 0.7], fontsize=26)
ax.xaxis.get_offset_text().set_fontsize(26)

plt.grid(lw=0.5, ls=":")
plt.tight_layout()
plot_name = f"privacy_iid.pdf"
plt.savefig(plot_name, bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```
