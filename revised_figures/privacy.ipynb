```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import rc
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import ScalarFormatter
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
# title = r'$\mathrm{Synthetic~(2~obervations)}$'
noise_ratio = [0, 0.0025, 0.005, 0.0075, 0.01]

n5000 = [0.0082952, 0.0611289, 0.1225977, 0.1599407, 0.1977497]
n5000_std = [0.0000380, 0.0036398, 0.0153141, 0.0294529, 0.0363449]

n10000 = [0.00748573, 0.0415048, 0.0917367, 0.1249944, 0.1522843]
n10000_std = [0.0000325, 0.0072291, 0.0166303, 0.0161943, 0.0292621]

n20000 = [0.00748573, 0.0283594, 0.0638198, 0.1074218, 0.1388770]
n20000_std = [0.0000325, 0.0072291, 0.0166303, 0.0098723, 0.0334883]

coeffs = np.polyfit(noise_ratio, n5000, 1)
slope, intercept = coeffs
print(f"slope for n5000: {slope}")
print(f"dis: {intercept}")

coeffs = np.polyfit(noise_ratio, n10000, 1)
slope, intercept = coeffs
print(f"slope for n10000: {slope}")
print(f"dis: {intercept}")
coeffs = np.polyfit(noise_ratio, n20000, 1)
slope, intercept = coeffs
print(f"slope for n20000: {slope}")
print(f"dis: {intercept}")

f, ax = plt.subplots(figsize=(10,8))

#enhance = ob2_rmse_mean / rmse_mean
#enhance = rmse_mean / ob2_rmse_mean
#print(enhance.mean())
#print(ob2_err_mean)
#print(err_mean)

plt.plot(noise_ratio, n5000, lw=5, color='royalblue',ls="dashdot" )
plt.plot(noise_ratio, n10000, lw=5, color='gold', ls="dotted")
plt.plot(noise_ratio, n20000, lw=5, color='red', ls="dashed")

plt.scatter(noise_ratio, n5000, s=100, color='royalblue', label=r"$n=5d$", marker="X")
plt.scatter(noise_ratio, n10000, s=100, color='gold', label=r"$n=10d$")
plt.scatter(noise_ratio, n20000, s=100, color='red', label=r"$n=20d$", marker="D")

plt.errorbar(noise_ratio, n5000, yerr=n5000_std,color='royalblue', fmt='X', capsize=10)
plt.errorbar(noise_ratio, n10000, yerr=n10000_std,color='gold', capsize=10)
plt.errorbar(noise_ratio, n20000, yerr=n20000_std,color='red', fmt='D', capsize=10)

#plt.legend(fontsize=26, loc='center left',bbox_to_anchor=(1, 0.5), ncol=1)
plt.legend(fontsize=38, loc='upper right')

plt.xlabel(r"$\sigma$", fontsize=40)
plt.ylabel(r"$\mathrm{Estimation~Error}$", fontsize=40)


xticks = [0, 0.0025, 0.005, 0.0075, 0.01]
plt.xticks(xticks, fontsize=40)
plt.ticklabel_format(style='sci', scilimits=(0,0), axis='x')
plt.yticks([0, 0.1, 0.2], fontsize=40)
ax.xaxis.get_offset_text().set_fontsize(40)
plt.gca().invert_xaxis()

plt.grid(lw=0.5, ls=":")
plt.tight_layout()
plot_name = f"privacy_n.pdf"
plt.savefig(plot_name, bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```
