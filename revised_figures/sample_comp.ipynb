```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import rc
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
# title = r'$\mathrm{Synthetic~(2~obervations)}$'
d2 = [1000, 2000, 3000, 4000, 5000]

m20 = [0.011217404156923294, 0.011327185668051243, 0.01230093278, 0.0112696588, 0.01238064095377922]
m50 = [0.00867448654, 0.007825225592, 0.008368478157, 0.009209289216, 0.00835324917]
m100 = [0.00618130248, 0.005849636625, 0.00554137351, 0.005278910976, 0.005322077777]

f, ax = plt.subplots(figsize=(8,7))

#enhance = ob2_rmse_mean / rmse_mean
#enhance = rmse_mean / ob2_rmse_mean
#print(enhance.mean())
#print(ob2_err_mean)
#print(err_mean)

plt.plot(d2, m20, lw=3, color='royalblue',ls="dashdot" )
plt.plot(d2, m50, lw=3, color='gold', ls="dotted")
plt.plot(d2, m100, lw=3, color='red', ls="dashed")

plt.scatter(d2, m20, s=100, color='royalblue', label=r"$m=20d_2$", marker="X")
plt.scatter(d2, m50, s=100, color='gold', label=r"$m=50d_2$")
plt.scatter(d2, m100, s=100, color='red', label=r"$m=100d_2$", marker="D")

plt.legend(fontsize=26, loc='upper right', bbox_to_anchor=(1.02, 0.44))

plt.xlabel(r"$d_2$", fontsize=40)
plt.ylabel(r"$\mathrm{Estimation~Error}$", fontsize=40)

xticks = d2
plt.xticks(xticks, fontsize=40)
plt.ticklabel_format(style='sci', scilimits=(0,0), axis='x')
plt.yticks([0.005, 0.007, 0.009, 0.011, 0.013], fontsize=40)
ax.xaxis.get_offset_text().set_fontsize(32)

plt.grid(lw=0.5, ls=":")
plt.tight_layout()
plot_name = f"sample_comp_iid.pdf"
plt.savefig(plot_name, bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```
