```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import rc
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator, NullFormatter, ScalarFormatter

mpl.rcParams.update(mpl.rcParamsDefault)

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed
```

```python
T_hat_bias = [2e-3, 6e-4, 3e-4, 2e-3, 6e-4, 2.7e-4]
categories = ['$p=0.01$', '$p=0.05$', '$p=0.1$', '$C=10$', '$C=50$', '$C=100$']
bar_width = 0.35
x = np.arange(len(categories)) * 1.5

fig, ax = plt.subplots(figsize=(15, 6))
ax.bar(x, T_hat_bias, width=bar_width, color='royalblue', label=r'$\widehat{T}$')

ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=36, rotation=10)

plt.ylabel(r"$\mathrm{Bias}$", fontsize=36)
plt.yscale('log')
#plt.yticks([1e-3],fontsize=36)
ax.yaxis.set_major_formatter(NullFormatter()) 
ax.yaxis.set_minor_formatter(NullFormatter())
plt.ylim(0.99e-4, 1.01e-2)
ax.set_yticks([1e-4, 1e-3, 1e-2])
ax.set_yticklabels([r'$10^{-4}$', r'$10^{-3}$', r'$10^{-2}$'], fontsize=36)

ax.grid(axis='y', which='major', linestyle='--', linewidth=1, color='gray', alpha=0.6)

plt.legend(fontsize=36)
#plt.title(r"$\mathrm{Bias}$", fontsize=36)

plt.tight_layout()


plt.savefig('T_hat_bias.pdf', bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```

```python
T_hat_var = [1e-10, 5.7e-12, 1.4e-12, 1.1e-10, 5.8e-12, 1.4e-12]
T_bar_var = [1e-5, 6.3e-7, 1.6e-7, 9.2e-6, 6.1e-7, 1.5e-7]
categories = ['$p=0.01$', '$p=0.05$', '$p=0.1$', '$C=10$', '$C=50$', '$C=100$']

bar_width = 0.35
bar_offset = 0.03

x = np.arange(len(categories)) * 1.5

fig, ax = plt.subplots(figsize=(15, 6))
ax.bar(x - bar_width/2-bar_offset, T_hat_var, width=bar_width, color='royalblue', label=r'$\widehat{T}$')
ax.bar(x + bar_width/2+bar_offset, T_bar_var, width=bar_width, color='orange', label=r'$\overline{T}$', hatch='/')


ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=36, rotation=10)

plt.ylabel(r"$\mathrm{Variance}$", fontsize=40)
plt.yscale('log')
plt.yticks(fontsize=36)
#ax.yaxis.set_major_formatter(NullFormatter()) 
#ax.yaxis.set_minor_formatter(NullFormatter())
plt.ylim(1e-13, 1e-1)
ax.set_yticks([1e-12, 1e-9, 1e-6, 1e-3])
ax.set_yticklabels([r'$10^{-12}$', r'$10^{-9}$', r'$10^{-6}$', r'$10^{-3}$'], fontsize=36)

plt.legend(fontsize=36, ncol=2)
#plt.title(r"$\mathrm{Variance}$", fontsize=36)

ax.grid(axis='y', which='major', linestyle='--', linewidth=1, color='gray', alpha=0.6)


plt.tight_layout()

plt.savefig('T_hat_var.pdf', bbox_inches='tight', format="pdf", dpi=1200)

plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(16, 6))
x = np.arange(9)
y = np.random.rand(9) * 1e-3
ax.bar(x, y)
ax.set_yscale('log')

# 设置 x tick
categories = [f'Cat{i}' for i in range(9)]
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, fontsize=12)

# transAxes 横线（下面标注）
x0_axes, x1_axes = 0.65, 0.95
line_y = -0.15
text_y = -0.22
from matplotlib.lines import Line2D

## 样式参数
bracket_color = 'dimgray'
bracket_width = 1.5
vline_height = 0.025

# 横线
line = Line2D([x0_axes, x1_axes], [line_y, line_y], transform=ax.transAxes,
              color=bracket_color, linewidth=bracket_width, clip_on=False)
ax.add_line(line)

# 左竖线
vline_left = Line2D([x0_axes, x0_axes], [line_y, line_y + vline_height], transform=ax.transAxes,
                    color=bracket_color, linewidth=bracket_width, clip_on=False)
ax.add_line(vline_left)

# 右竖线
vline_right = Line2D([x1_axes, x1_axes], [line_y, line_y + vline_height], transform=ax.transAxes,
                     color=bracket_color, linewidth=bracket_width, clip_on=False)
ax.add_line(vline_right)
ax.text((x0_axes+x1_axes)/2, text_y, 'Group Label', transform=ax.transAxes, ha='center', fontsize=14)

# 关键：保留底部区域显示
fig.subplots_adjust(bottom=0.3)

plt.show()
```
