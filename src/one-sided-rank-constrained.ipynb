try svd solution

```python
sym_M = observed_M.T @ observed_M

diag_M = np.diag(np.diag(sym_M))

# reweight diagonal vs off-diagonal entries

reweighted_M = (1.0 / p) * diag_M + (1.0 / (p * p)) * (sym_M - diag_M)

sym_U, sym_D, sym_Vt = np.linalg.svd(reweighted_M)

# cut down to rank

rank_reduced_M = sym_Vt[:rank, :].T @ sym_Vt[:rank, :]
```

```python
U, D, Vt = np.linalg.svd(M.T @ M)

print(sym_D)

print(D)

print(rank)
```

```python
right_subspace_matrix = Vt[:rank, ].T @ Vt[: rank, :]

err_matrix = right_subspace_matrix - rank_reduced_M

err = np.linalg.norm(err_matrix, 'fro')

right_norm = np.linalg.norm(right_subspace_matrix, 'fro')

print(err, right_norm)
```
