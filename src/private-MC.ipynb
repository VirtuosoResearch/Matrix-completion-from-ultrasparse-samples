```python
import numpy as np
import random
from data import get_random_matrix, get_uniformly_random_samples
from gradient_descent import symmetric_noisy_gradient_descent
from nuclear_norm_min import nuclear_norm_minimization
```

```python
# instance parameters
d = 100
r = 5
p = 5 * d * r / (d**2)
```

```python
U = get_random_matrix(d, d, r)

A = U @ U.T

values, S = get_uniformly_random_samples(A, p)

non_zero_indices = np.nonzero(np.array(S))

#print(non_zero_indices)
```

set up noisy gradient descent update

```python
eta = 0.001
epochs = 500
noise_var = 5.0
reg = 0.0
init_scale = 0.01

X = symmetric_noisy_gradient_descent(values, S, d, eta, epochs, noise_var, reg, init_scale)

dist = np.linalg.norm(X @ X.T - A, 'fro') / np.linalg.norm(A, 'fro')
print(dist)
```

with regularization

```python
reg = 10.00

X_reg = symmetric_noisy_gradient_descent(values, S, d, eta, epochs, noise_var, reg)

dist = np.linalg.norm(X_reg @ X_reg.T - A, 'fro') / np.linalg.norm(A, 'fro')
print(dist)
```

nuclear norm minimization

```python
noise_var = 5
perturbation = np.random.normal(0, 1, (d, d)) * noise_var
zero_matrix = np.zeros_like(perturbation)
zero_matrix[non_zero_indices] = perturbation[non_zero_indices]

perturbed_values = values +zero_matrix
```

```python
X_nuc, nuc_val = nuclear_norm_minimization(perturbed_values, non_zero_indices)
```

```python
dist = np.linalg.norm(X_nuc - A, 'fro') / np.linalg.norm(A, 'fro')
print(dist)
```
