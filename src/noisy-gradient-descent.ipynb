```python
import numpy as np
import random as random
```

```python
def get_random_matrix(d1, d2, r):
	X = np.random.normal(0, 1, size = (d1, d2))

	U, D, Vt = np.linalg.svd(X, full_matrices=False)
	D[r:] = 0
	X = U @ np.diag(D) @ Vt
	return X
```

```python
d1 = 20
d2 = 15
r = 5
```

```python
X = get_random_matrix(d1, d2, r)
```
